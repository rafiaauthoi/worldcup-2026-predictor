```python
from statsbombpy import sb
import pandas as pd

# See what competitions StatsBomb has for free
competitions = sb.competitions()
print(competitions.shape)
competitions.head(20)
```

```python
# Filter for World Cup competitions
wc = competitions[competitions['competition_name'] == 'FIFA World Cup']
print(wc[['competition_id', 'season_id', 'season_name', 'country_name']])
```

```python
# Get all 2022 World Cup matches
matches = sb.matches(competition_id=43, season_id=106)
print("Total matches:", len(matches))
print("\nColumns:", matches.columns.tolist())
matches.head()
```

```python
# Get events for one match to see what player data looks like
sample_match_id = matches['match_id'].iloc[0]
events = sb.events(match_id=sample_match_id)

print("Total events in this match:", len(events))
print("\nEvent types available:")
print(events['type'].value_counts())
```

```python
# See what player info and columns we have
print("Columns:", events.columns.tolist())
print("\nSample shot event:")
shots = events[events['type'] == 'Shot']
print(shots[['player', 'team', 'shot_outcome', 'shot_statsbomb_xg']].head(10))
```

```python
import warnings
warnings.filterwarnings('ignore')

all_events = []

print("Fetching events for all 64 matches...")
for i, match_id in enumerate(matches['match_id']):
    events = sb.events(match_id=match_id)
    all_events.append(events)
    if (i+1) % 10 == 0:
        print(f"  {i+1}/64 done")

all_events_df = pd.concat(all_events, ignore_index=True)
print(f"\nTotal events: {len(all_events_df)}")
print("Done!")
```

```python
# Aggregate player stats across all WC 2022 matches

# Goals and shots
shots = all_events_df[all_events_df['type'] == 'Shot'].copy()
player_goals = shots[shots['shot_outcome'] == 'Goal'].groupby(['player', 'team']).size().reset_index(name='goals')
player_shots = shots.groupby(['player', 'team']).size().reset_index(name='shots')
player_xg = shots.groupby(['player', 'team'])['shot_statsbomb_xg'].sum().reset_index(name='xg')

# Yellow cards
fouls = all_events_df[all_events_df['type'] == 'Foul Committed'].copy()
yellow_cards = fouls[fouls['foul_committed_card'] == 'Yellow Card'].groupby(['player', 'team']).size().reset_index(name='yellow_cards')
red_cards = fouls[fouls['foul_committed_card'].isin(['Red Card', 'Second Yellow'])].groupby(['player', 'team']).size().reset_index(name='red_cards')

# Passes
passes = all_events_df[all_events_df['type'] == 'Pass'].copy()
player_passes = passes.groupby(['player', 'team']).size().reset_index(name='passes')
pass_assists = passes[passes['pass_goal_assist'] == True].groupby(['player', 'team']).size().reset_index(name='assists')

# Merge everything
from functools import reduce
dfs = [player_shots, player_goals, player_xg, player_passes, pass_assists, yellow_cards, red_cards]
player_stats = reduce(lambda left, right: pd.merge(left, right, on=['player', 'team'], how='outer'), dfs)
player_stats = player_stats.fillna(0)

# Convert to int where appropriate
for col in ['goals', 'shots', 'passes', 'assists', 'yellow_cards', 'red_cards']:
    player_stats[col] = player_stats[col].astype(int)

player_stats = player_stats.sort_values('goals', ascending=False).reset_index(drop=True)

print("Player stats shape:", player_stats.shape)
print("\nTop 10 scorers:")
print(player_stats[['player', 'team', 'goals', 'shots', 'xg', 'assists']].head(10))
```

```python
# Save player stats
player_stats.to_csv('data/processed/player_stats_wc2022.csv', index=False)
print("Player stats saved!")

# Build team-level features from player stats
team_stats = player_stats.groupby('team').agg(
    total_goals=('goals', 'sum'),
    total_shots=('shots', 'sum'),
    total_xg=('xg', 'sum'),
    total_passes=('passes', 'sum'),
    total_assists=('assists', 'sum'),
    total_yellow_cards=('yellow_cards', 'sum'),
    total_red_cards=('red_cards', 'sum'),
    top_scorer_goals=('goals', 'max'),
    top_scorer_xg=('xg', 'max'),
    squad_size=('player', 'count')
).reset_index()

team_stats = team_stats.sort_values('total_goals', ascending=False).reset_index(drop=True)

print("\nTeam stats shape:", team_stats.shape)
print("\nTop 10 teams by goals:")
print(team_stats[['team', 'total_goals', 'total_xg', 'total_yellow_cards', 'top_scorer_goals']].head(10))

# Save team stats
team_stats.to_csv('data/processed/team_stats_wc2022.csv', index=False)
print("\nTeam stats saved!")
```
